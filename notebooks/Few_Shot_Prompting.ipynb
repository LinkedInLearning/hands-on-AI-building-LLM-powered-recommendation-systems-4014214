```python
# 1. Import necessary libraries and load the model
from transformers import pipeline

# This might take a moment to download the model weights
generator = pipeline('text-generation', model='gpt2-medium')

# 2. Define the ZERO-SHOT prompt
prompt_zero_shot = """ Task: Write a justification for the following recommendation. Context: The user enjoys the movie "The Matrix".
Recommendation: Suggest the movie "Blade Runner". Justification: """

# 3. Generate the response
response = generator( prompt_zero_shot,
                     max_new_tokens=30, # Controls the length of the new text
                    num_return_sequences=1 )
# 4. Print the generated part of the text
print(response[0]['generated_text'][len(prompt_zero_shot):])
```

```python
# 1. Define the FEW-SHOT prompt with examples
prompt_few_shot = """
# Example 1
Context: A user likes classic sci-fi novels with deep philosophical themes.
Recommendation: "Dune"
Explanation: Since you enjoy sci-fi that explores complex ideas like politics and destiny, you'll love the intricate world-building of "Dune".

# Example 2
Context: A user recently watched several fast-paced action movies.
Recommendation: "John Wick"
Explanation: Based on your interest in high-octane action and slick choreography, "John Wick" is a perfect next watch.

# Our Actual Task
Context: The user enjoys the movie "The Matrix".
Recommendation: "Blade Runner"
Explanation: """

# 2. Generate the response
response = generator( prompt_few_shot, max_new_tokens=40, num_return_sequences=1 )

# 3. Print the generated part of the text
print(response[0]['generated_text'][len(prompt_few_shot):])
```
