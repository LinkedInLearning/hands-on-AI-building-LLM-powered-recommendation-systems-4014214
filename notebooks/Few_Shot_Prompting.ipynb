```python
# Import necessary libraries and load the model
from transformers import pipeline

# Download the model weights
generator = pipeline('text-generation', model='gpt2-medium')
```

```python
# 1. Define the ZERO-SHOT prompt
prompt_zero_shot = """
Task: Write a justification for the following movie recommendation given the respective context. Don't output anything else.

Context: The user enjoys the movie "The Matrix".
Recommendation: Suggest the movie "Blade Runner".
Justification: """

# 2. Generate the response
response = generator(
              prompt_zero_shot,
              max_new_tokens=30, # Controls the length of the new text
              num_return_sequences=1
           )
# 3. Print the generated part of the text
print(response[0]['generated_text'][len(prompt_zero_shot):])
```

```python
# 1. Define the FEW-SHOT prompt with examples
prompt_few_shot = """
Here are two examples for user context, movie recommendation, and justification of why movie was recommended.

Context: A user likes classic sci-fi novels with deep philosophical themes.
Recommendation: "Dune"
Justification: Since you enjoy sci-fi that explores complex ideas like politics and destiny, you'll love the intricate world-building of "Dune".

Context: A user recently watched several fast-paced action movies.
Recommendation: "John Wick"
Justification: Based on your interest in high-octane action and slick choreography, "John Wick" is a perfect next watch.

====
Task: Write a justification for the following movie recommendation given the respective context. Don't output anything else.
Context: The user enjoys the movie "The Matrix".
Recommendation: "Blade Runner"
Justification: """

# 2. Generate the response
response = generator( prompt_few_shot, max_new_tokens=40, num_return_sequences=1 )

# 3. Print the generated part of the text
print(response[0]['generated_text'][len(prompt_few_shot):])
```
