```python
from sentence_transformers import SentenceTransformer, util

# 1. Load a pre-trained model. 'all-MiniLM-L6-v2' is a great, fast starting point.
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Define some product descriptions we want to convert.
product_descriptions = [
    "A sleek, modern smartphone with a stunning OLED display and 12-hour battery life.",
    "The latest mobile phone featuring a vibrant screen and all-day power.",
    "A classic, hardcover novel about a detective solving a mystery in 1940s London.",
    "A thrilling mystery book set in the gritty streets of post-war London."
]
```

```python
# 3. Use the model to create the embeddings.
embeddings = model.encode(product_descriptions)

# Let's see what we got.
for description, embedding in zip(product_descriptions, embeddings):
    print("Description:", description)
    # We'll just print the first 3 numbers of the vector to see what it looks like.
    print("Embedding shape:", embedding.shape)
    print("Embedding snippet:", embedding[:3])
    print("-" * 20)
```

```python
# 4. Let's compute the similarity between the first two descriptions.
cosine_scores = util.cos_sim(embeddings[0], embeddings[1])
print(f"Similarity between phone descriptions: {cosine_scores[0][0]:.2f}")

# 5. Now let's compare a phone description to a book description.
cosine_scores_diff = util.cos_sim(embeddings[0], embeddings[2])
print(f"Similarity between phone and book: {cosine_scores_diff[0][0]:.2f}")
```
