```python
# Import necessary libraries and load the model
from transformers import pipeline

# Download the model weights
generator = pipeline('text-generation', model='gpt2-medium')
```

```python
# 1. Basic prompt
basic_prompt = """
User viewed:
* Men's Hawaiian Wrinkle Free Short Sleeve Button Down Shirt
* Funky Hawaiian Shirt for Men Palm Beach Shirt
* COOFANDY Mens Button Down Short Sleeve Shirt Casual Shirts Summer

Question: What are they interested in?

Answer:
"""

# 2. Generate the response
response = generator(
              basic_prompt,
              max_new_tokens=50, # Controls the length of the new text
              num_return_sequences=1
           )
# 3. Print the generated part of the text
print(response[0]['generated_text'][len(basic_prompt):])
```

```python
# 1. Define the quality prompt
quality_prompt = """
Role: You are an expert personal shopping assisstant.

Context: A client has recently viewed the following items:
* Men's Hawaiian Wrinkle Free Short Sleeve Button Down Shirt
* Funky Hawaiian Shirt for Men Palm Beach Shirt
* COOFANDY Mens Button Down Short Sleeve Shirt Casual Shirts Summer

Task: Based on this entire history, please generate a 2-sentence summary of the client's current inferred interests and shopping goals.
Focus on themes, styles, and potential use cases.
"""

# 2. Generate the response
response = generator( quality_prompt, max_new_tokens=50, num_return_sequences=1 )

# 3. Print the generated part of the text
print(response[0]['generated_text'][len(quality_prompt):])
```
